# Basic plots

```python
# import libraries needed for plotting
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# define a python function for maths function
def f(x):
    return x
#generate the values of x into a variable 
x=np.arange(0,10,.1)

#get the values of y from the function
y = f(x)
```

```python
# Plottting a basic line plot using matplotlib
plt.figure()
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# plot a basic line plot using plot express
px.line(x,y, labels = {"x":"x-axis","index": "y axis"})
```

```python
# define a function to plot a parabola
def f(x):
    return np.power(x,2)

x2=np.arange(-10000,10000, 1)
y = f(x2)
```

```python
# plot  a parabola plot using plot express
px.line(x=x2,y =y, labels = {"x":"x-axis","index": "y axis"})
```

```python
# plot a parabola plot using matplotlib
plt.figure()
plt.plot(x2,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# define an exponential function
def f(x):
    return np.exp(x)
    
x3=np.arange(0,10,0.01)
y =f(x3)

# plot an exponential function using plot express
px.line(x=x3,y =y, labels = {"x":"x-axis","index": "y axis"})
```

```python
# define a logistical regression function
def f(x):
    return 1/(1+np.exp(-x))

x4=np.arange(-10,10,0.01)
y = f(x4)

# plot the logistical regression function using plot express
px.line(x=x4,y =y, labels = {"x":"x-axis","index": "y axis"})



    
```
